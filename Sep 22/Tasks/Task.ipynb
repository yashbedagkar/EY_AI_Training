```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
plt.bar(df["Month"],df["Revenue"],color="skyblue")
plt.title("Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.bar(df["Month"],df["Units_Sold"],color="skyblue")
plt.title("Units_Sold by Month")
plt.xlabel("Month")
plt.ylabel("Units_Sold")
plt.show()
```

```python
plt.plot(df["Month"],df["Units_Sold"],color="skyblue")
plt.title("Units_Sold by Month")
plt.xlabel("Month")
plt.ylabel("Units_Sold")
plt.show()
```
