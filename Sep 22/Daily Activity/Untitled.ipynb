```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Rahul", "Priya", "Arjun", "Neha", "Vikram"],
    "Age": [21, 22, 20, 23, 21],
    "Course": ["AI", "ML", "Data Science", "AI", "ML"],
    "Marks": [85, 90, 78, 88, 95]
}

df= pd.DataFrame(data)
```

```python
df
```

```python
print(df["Name"])
print(df[["Name","Age"]])
print(df.iloc[0])
print(df.loc[2,"Course"])
```

```python
high_scores= df[df["Marks"] > 85]
print(high_scores)
```

```python
df["Result"] = np.where(df["Marks"] >= 80, "Pass", "Fail")
```

```python
import matplotlib.pyplot as plt
```

```python
avg_marks = df.groupby("Course")["Marks"].mean()
```

```python
avg_marks.plot(kind="bar", color="skyblue",edgecolor="black")
plt.title("Avg Marks by course")
plt.xlabel("Course")
plt.ylabel("Avg marks")
plt.xticks(rotation=45)
plt.show()
```
